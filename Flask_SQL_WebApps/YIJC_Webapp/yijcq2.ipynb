```python
#Task 2.1

import sqlite3

create_studentdb = """
CREATE TABLE IF NOT EXISTS `Student` (
	`MatricNo`	INTEGER,
	`Name`	TEXT,
	`Gender`	TEXT,
	`CivicsClass`	TEXT,
	PRIMARY KEY(`MatricNo`),
    FOREIGN KEY(`CivicsClass`) REFERENCES Civics(`Class`)
);

"""

create_civicsdb = """
CREATE TABLE IF NOT EXISTS `Civics` (
	`Class`	TEXT,
	`Tutor`	TEXT,
	`HomeRoom`	TEXT,
	PRIMARY KEY(`Class`)
);
"""
create_ccainfo = """
CREATE TABLE IF NOT EXISTS `CCAInfo` (
	`Name`	TEXT,
	`TeacherIC`	TEXT,
    PRIMARY KEY(`Name`)
);

"""

create_studentcca = """
CREATE TABLE IF NOT EXISTS `StudentCCA` (
	`MatricNo`	INTEGER,
	`CCAName`	TEXT,
	FOREIGN KEY(`MatricNo`) REFERENCES Student(`MatricNo`),
    FOREIGN KEY(`CCAName`) REFERENCES CCAInfo(`Name`)
);
"""

conn = sqlite3.connect('student2.db')
cursor = conn.cursor()

cursor.execute(create_studentdb)
cursor.execute(create_civicsdb)
cursor.execute(create_ccainfo)
cursor.execute(create_studentcca)

conn.commit()
```

```python
#Task 2.2
import csv

with open('Student.csv') as f:
    csvreader = csv.reader(f, delimiter=",")
    next(csvreader)
    for row in csvreader:
        print(row)
        cursor.execute("INSERT INTO STUDENT(MatricNo, Name, Gender, CivicsClass) VALUES(?,?,?,?)",(row[0],row[1],row[2],row[3]))
        
    conn.commit()
    
with open('Civics.csv') as f:
    csvreader = csv.reader(f, delimiter=",")
    next(csvreader)
    for row in csvreader:
        print(row)
        cursor.execute("INSERT INTO Civics(Class, Tutor, HomeRoom) VALUES(?,?,?)",(row[0],row[1],row[2]))
    
    conn.commit()
    
    
with open('CCAInfo.csv') as f:
    csvreader = csv.reader(f, delimiter=",")
    next(csvreader)
    for row in csvreader:
        print(row)
        cursor.execute("INSERT INTO CCAInfo(Name, TeacherIC) VALUES(?,?)",(row[0],row[1]))
    conn.commit()
    
    
with open('StudentCCA.csv') as f:
    csvreader = csv.reader(f, delimiter=",")
    next(csvreader)
    for row in csvreader:
        print(row)
        cursor.execute("INSERT INTO StudentCCA(MatricNo, CCAName) VALUES(?,?)",(row[0],row[1]))
    conn.commit()
```
