```
# repeat the process, removing the end number at each iteration, until the end
import random
import time
import matplotlib.pyplot as plt


def bubbleSort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n-i-1):

            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    print(arr)

arr = [68,34,12,5,7,90,83,45,46]
bubbleSort(arr)

testarr = []
testcases = 10
numtestcasey = []
timingsx = []
for l in range(20):
    a = time.time()
    for k in range(testcases):
        testarr.append(random.randint(0,testcases))
    bubbleSort(testarr)
    b = time.time()
    timingsx.append(b-a)
    testcases = testcases+50
    numtestcasey.append(testcases)

#time complexity
plt.plot(numtestcasey,timingsx)
plt.title('The time clomplexity of a bubble_sort_Algorithm')
plt.xlabel('No. of test cases')
plt.ylabel('Time required (s)')

plt.show()
plt.scatter(numtestcasey,timingsx)
plt.show()
print(numtestcasey)
print(timingsx)
```
